### II. `sklearn` to the rescue üöë

Most of the time, you don't need to reinvent the wheel, and you can directly use an existing implementation of the function you are looking for. It might be within `scikit-learn`, but also in any other library.

That's part of the magic of Python. ‚ú®

In this particular case for example, `sklearn` contains a ready-to-use function `train_test_split` that you will use now.

**Q4**. Import `train_test_split` function from `sklearn` library and call it in order to split X and y variables.

Do you need to specify *shuffle* argument? 

Extra points if you can split X and y data in the same line.

```python
# TODO: import train_test_split from Scikit-Learn
from sklearn.model_selection import train_test_split
```

```python
# TODO: use `train_test_split` on X and y with a test_size=.5 
# in order to create X_train, X_test

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.5, random_state=42)
```
